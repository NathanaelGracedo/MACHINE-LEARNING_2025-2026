## Import Library 

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Pra Pengolahan Data

Pra Pengolahan Data Training

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('dataset/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

Pra Pengolahan Data Testing

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('dataset/test_set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

## Pembuatan Model CNN

Inisiasi Model CNN

```python
cnn = tf.keras.models.Sequential()
```

Pembuatan Layer Konvulsi 1

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

Pembuatan Layer Pooling 1

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

Pembuatan Layer Konvulsi 2 dan Pooling 2

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

Flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

Fully Connected Layer 1 (input)

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

Fully Connecteed Layer 2 (output)

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Compile Model CNN

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## Fit CNN

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

Prediksi dengan 1 Citra

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/single_prediction/cat_or_dog_1.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'

print(prediction)
```
