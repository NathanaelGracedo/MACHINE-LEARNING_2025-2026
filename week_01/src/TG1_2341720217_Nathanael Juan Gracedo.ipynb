Jawab:

**Dampak Negatif Limbah Elektronik (E-Waste) dari AI:**
1. **Siklus Upgrade Cepat**: Inovasi AI yang cepat memaksa pergantian konstan perangkat keras (GPU/TPU) yang masih layak pakai, menghasilkan volume e-waste besar.
2. **Desain Tertutup & Tidak Modular**: Perangkat keras AI khusus sering dirancang sebagai satu unit, menyulitkan perbaikan atau upgrade sebagian, sehingga memperbanyak limbah.
3. **Eksploitasi Sumber Daya & Polusi**: Produksi dan pembuangan hardware AI menghabiskan logam langka, mineral konflik, dan mengandung bahan beracun yang mencemari lingkungan jika tidak didaur ulang dengan benar.

**Cara Mengatasi**:
1. **Ekonomi Sirkular**: Menerapkan desain hardware yang dapat diperbaiki/di-upgrade, serta program take-back untuk remanufaktur atau kanibalisasi bagian yang masih berguna.
2. **Perpanjang Masa Pakai**: Mengoptimalkan penggunaan hardware lama dengan algoritma AI yang efisien dan memprioritaskan komputasi cloud yang terpusat untuk memaksimalkan utilisasi.
3. **Prioritas pada Efisiensi**: Mengembangkan algoritma yang ringan (model distillation, quantization) sehingga tidak selalu membutuhkan hardware terbaru, mengurangi tekanan untuk upgrade.
4. **Regulasi & Transparansi**: Menerapkan regulasi tanggung jawab produsen (Extended Producer Responsibility/EPR) dan mendorong pelaporan jejak e-waste secara transparan.

**Referensi**
1. **The Shift Project. (2019)**. Lean ICT: Towards Digital Sobriety.
Menganjurkan "kesederhanaan digital" dan memperpanjang umur perangkat untuk mengurangi e-waste.
2. **Brevini, B. (2022)**. Is AI Good for the Planet?. Polity Press.
Menganalisis dampak material dan siklus hidup perangkat keras AI yang menyebabkan e-waste.
